```python
"""
    Diseñe un algoritmo que determine el resultado de la elección del representante estudiantil de la
    universidad X, para ello se presentaron tres candidatos A, B, y C.
    Para ganar la elección se debe obtener como mínimo el 51%.
    En caso que no haya un ganador se repite la elección en una segunda vuelta.
    Van a la segunda vuelta los dos candidatos que obtengan la más alta votación.
    Se anula la elección en caso de producirse un empate doble por el segundo lugar o un empate triple.
"""

#se importa la libreria random para sacar la funcion randrange y obtener numeros aleatorios
from random import randrange

#acumuladores para el ganador y candidato A,B,C
ganador = 0
contA = 0
contB = 0
contC = 0

#ciclo que se ejecuta hasta que encuentra un ganador 
while ganador != 1:

    #ciclo que simulara si 100 personas votaran
    for i in range(100):

        #voto aleatorio entre 1 y 3
        vote = randrange(1, 4)

        #condicion donde se le asigna un numero a cada candidato
        #si el numero es 1, 2 o 3, se le acumulara a cada candidato el voto
        if vote == 1:
            contA += 1
        elif vote == 2:
            contB += 1
        elif vote == 3:
            contC += 1

    #condicion para saber si alguno gana en primera vuelta
    if contA > contB and contA > contC:
        print("Gano A")

        #en cuanto encuentra un ganador, termina el ciclo 
        ganador = 1
    elif contB > contA and contB > contC:
        print("Gano B")
        ganador = 1
    elif contC > contB and contC > contA:
        print("Gano C")
        ganador = 1

    #condicion para saber si A y B quedan en empate
    elif contA == contB:
        print("Empate A y B")
        print("Ejecutando segunda vuelta")

        #Los contadores se reinician para que no afencte la votacion
        contA = 0
        contB = 0

        #re repite la votacion aleatorio ahora entre 1 y 2 para decidir entre dos candidatos
        for i in range(100):
            vote = randrange(1, 3)
            if vote == 1:
                contA += 1
            elif vote == 2:
                contB += 1
                
##condicion para saber quien gana entre A y B
#El proceso se repite para los demas empates
        if contA > contB:
            print("Gano A")
            ganador = 1
        elif contB > contA:
            print("Gano B")
            ganador = 1
    elif contA == contC:
        print("Empate A y C")
        print("Ejecutando segunda vuelta")
        contA = 0
        contC = 0
        for i in range(100):
            vote = randrange(1, 3)
            if vote == 1:
                contA += 1
            elif vote == 2:
                contC += 1
        if contA > contC:
            print("Gano A")
            ganador = 1
        elif contC > contA:
            print("Gano C")
            ganador = 1
    elif contC == contB:
        print("Empate B y C")
        print("Ejecutando segunda vuelta")
        contB = 0
        contC = 0
        for i in range(100):
            vote = randrange(1, 3)
            if vote == 1:
                contB += 1
            elif vote == 2:
                contC += 1
        if contB > contC :
            print("Gano B")
            ganador = 1
        elif contC > contB:
            print("Gano C")
            ganador = 1

```
